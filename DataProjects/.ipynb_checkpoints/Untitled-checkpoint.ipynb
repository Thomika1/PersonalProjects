```python
import pandas as pd
import re
import matplotlib.pyplot as plt
from collections import Counter
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

```

```python
def FindAuthor(s):
    regex = r'^(.+?):'
    match = re.match(regex, s)
    if match:
        return match.group(1)
    return None

def getDataPoint(line):
   
    if not line.startswith('['):
        return None, None, None, None  
    
    # Divide a linha no primeiro '] ' para separar a data/hora do conteúdo
    splitLine = line.split('] ', 1)
    if len(splitLine) < 2:
        return None, None, None, None  
    
    dateTime = splitLine[0].strip('[')  
    # Divide a data e a hora
    if ', ' in dateTime:
        date, time = dateTime.split(', ', 1)  
    else:
        date, time = dateTime, None  
    
    content = splitLine[1].strip()  
    if ': ' in content:
        author = FindAuthor(content)
        message = content.split(': ', 1)[1]
    else:
        author = None
        message = content
    
    return date, time, author, message
```

```python
# Suponha que o arquivo de mensagens esteja no mesmo diretorio que o script
with open('_chat.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

data = []
for line in lines:
    date, time, author, message = getDataPoint(line)
    if date:  # Adiciona apenas linhas válidas (com data)
        data.append([date, time, author, message])

# Criar o DataFrame
df = pd.DataFrame(data, columns=['date', 'time', 'author', 'message'])

# Exibir o DataFrame
messages_per_day_per_user = df.groupby(['date', 'author']).size().reset_index(name='message_count')

total_messages_per_author = df.groupby('author').size().reset_index(name='total_messages')

# Ordenar por total de mensagens (opcional)
total_messages_per_author = total_messages_per_author.sort_values(by='total_messages', ascending=False)

# Exibir o resultado
# print(total_messages_per_author)
```

```python
# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(total_messages_per_author["author"], total_messages_per_author["total_messages"], color='skyblue')
plt.title('Total de Mensagens por Autor')
plt.xlabel('Autor')
plt.ylabel('Total de Mensagens')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.tight_layout()  # Ajusta o layout para evitar cortes
plt.show()
```

```python
# Converter a coluna 'date' para formato datetime
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y')

# Criar coluna de mês/ano para agrupar mensagens
df['month'] = df['date'].dt.to_period('M')

# Contar mensagens por mês e por usuário
messages_per_month_per_user = df.groupby(['month', 'author']).size().reset_index(name='message_count')

# Converter 'month' para string para facilitar no gráfico
messages_per_month_per_user['month'] = messages_per_month_per_user['month'].astype(str)

plt.figure(figsize=(12, 6))

messages_filtered = messages_per_month_per_user[messages_per_month_per_user['author'] != "Meta AI"]
messages_filtered = messages_filtered[messages_filtered['author'] != "Lucas Valdanha"]
messages_filtered = messages_filtered[messages_filtered['author'] != "reunião mundial"]




# Plotar uma linha para cada autor
for author in messages_filtered['author'].unique():
    user_data = messages_filtered[messages_filtered['author'] == author]
    plt.plot(user_data['month'], user_data['message_count'], marker='o', label=author)


# Melhorar formatação
plt.title('Quantidade de Mensagens por Mês para Cada Usuário')
plt.xlabel('Mês')
plt.ylabel('Número de Mensagens')
plt.xticks(rotation=45)  
plt.legend(title='Usuário', bbox_to_anchor=(1.05, 1), loc='upper left')  # Legenda fora do gráfico
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
# Lista de stopwords em português
stop_words = set(stopwords.words('portuguese'))

# Função para processar mensagens
def process_text(text):
    text = text.lower() 
    text = re.sub(r'[^\w\s]', '', text)  
    words = text.split()  
    words = [word for word in words if word not in stop_words and len(word) > 2]  # Remover stopwords e palavras curtas
    return words

# Criando um novo DataFrame com as palavras mais usadas
def contar_palavras(df):
    all_words = []
    
    # Aplicar a função para todas as mensagens
    df['message'].dropna().apply(lambda x: all_words.extend(process_text(x)))

    # Contar a frequência das palavras
    word_counts = Counter(all_words)

    # Criar um novo DataFrame ordenado
    df_words = pd.DataFrame(word_counts.items(), columns=['Palavra', 'Frequência'])
    df_words = df_words.sort_values(by='Frequência', ascending=False)

    return df_words



# Gerar o DataFrame com as palavras mais usadas
df_words = contar_palavras(df)

# Exibir o resultado
print(df_words.head(5))
```

```python
df_caga = df_words[df_words["Palavra"].str.startswith("caga")].reset_index(drop=True)
df_escaga = df_words[df_words["Palavra"].str.startswith("fezes")].reset_index(drop=True)

df_caga_total = pd.concat([df_caga, df_escaga]).reset_index(drop=True)
# Exibir o resultado
print(df_caga_total)

print(df_caga_total["Frequência"].sum())
```
