```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "/home/thomaz/Downloads/Mega-Sena.xlsx"
df = pd.read_excel(path)

```

```python
df.head()
```

```python
numeros = pd.concat([df["Bola1"],df["Bola2"],df["Bola3"],df["Bola4"],df["Bola5"],df["Bola6"]])
```

```python
plt.figure(figsize=(12, 6)) 
contagem = numeros.value_counts().sort_index()
contagem.plot(kind="bar")
contagem_df = contagem.reset_index()
contagem_df.columns = ['num','ocorrencias']

contagem_df.head()
```

```python
menores_seis = contagem_df.nsmallest(7, 'ocorrencias')
mediana = contagem_df.median()
media = contagem_df.mean()
print(media)
print(mediana)
menores_seis["comparacao mediana"] = menores_seis["ocorrencias"] - 282
menores_seis["comparacao media"] = menores_seis["ocorrencias"] - 280.9
menores_seis.head(7)
```
