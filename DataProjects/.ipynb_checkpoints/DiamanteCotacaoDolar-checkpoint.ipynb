```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta
```

```python
data_inicial = "2013-01-01"
data_final = "2023-01-01"


url = f"https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/CotacaoDolarPeriodo(dataInicial=@dataInicial,dataFinalCotacao=@dataFinalCotacao)?@dataInicial='01-01-2015'&@dataFinalCotacao='01-01-2025'&$top=10000&$format=json&$select=cotacaoVenda,dataHoraCotacao"
```

```python
# Requisio da API
response = requests.get(url)
dados = response.json()

# Converte os dados para um DataFrame
df = pd.DataFrame(dados["value"])

# Verifica se h dados
if df.empty:
    raise ValueError("Nenhum dado foi retornado pela API.")

# Extraindo apenas a parte da data e convertendo para datetime
df['data'] = pd.to_datetime(df['dataHoraCotacao'].str.split().str[0])

# Converte as datas para valores numricos (necessrio para np.polyfit)
df['data_num'] = mdates.date2num(df['data'])

# Converter data_final para datetime
data_final_dt = pd.to_datetime(data_final)

# Calcular a data de um ano atrs em relao  data_final
um_ano_atras = data_final_dt - timedelta(days=365)

# Filtrar o DataFrame para incluir apenas os dados do ltimo ano em relao  data_final
df_filtrado = df[(df['data'] >= um_ano_atras) & (df['data'] <= data_final_dt)]

# Verifica se h dados aps a filtragem
if df_filtrado.empty:
    raise ValueError("Nenhum dado encontrado no ltimo ano em relao  data_final.")

# Realizar a regresso linear nos dados filtrados
coef = np.polyfit(df_filtrado['data_num'], df_filtrado['cotacaoVenda'], 1)
poly1d_fn = np.poly1d(coef)

# Calcula o desvio padro da cotao de venda (apenas nos dados filtrados)
std_dev = df_filtrado['cotacaoVenda'].std()

# Criando o grfico
plt.figure(figsize=(20, 6))

# Plotando os pontos reais
plt.plot(df_filtrado['data'], df_filtrado['cotacaoVenda'], label="Cotao de Venda", marker='o', linestyle='-')

# Adicionando a linha de regresso
plt.plot(df_filtrado['data'], poly1d_fn(df_filtrado['data_num']), color='red', linestyle='--', label="Regresso Linear")

# Adicionando as linhas de desvio padro
plt.plot(df_filtrado['data'], poly1d_fn(df_filtrado['data_num']) + std_dev, color='green', linestyle='--', label="Desvio Padro Superior")
plt.plot(df_filtrado['data'], poly1d_fn(df_filtrado['data_num']) - std_dev, color='blue', linestyle='--', label="Desvio Padro Inferior")

# Formatando o eixo X para exibir os meses
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Mostrar por ms
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # Formato da data

# Adicionando ttulo e rtulos
plt.title(f'Cotao de Venda no ltimo Ano ({um_ano_atras.strftime("%Y-%m-%d")} a {data_final_dt.strftime("%Y-%m-%d")}) com Regresso Linear e Desvio Padro')
plt.xlabel('Data')
plt.ylabel('Cotao de Venda')
plt.legend()

# Exibindo o grfico
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Criando logica para contar quantas vezes a linha toca os desvios padrao 
print(df_filtrado.head())
```

```python
# Calcular os valores previstos da cotao de venda com base na regresso linear
valores_previstos = poly1d_fn(df_filtrado['data_num'])

# Calcular os limites dos desvios padro
limite_superior = valores_previstos + std_dev
limite_inferior = valores_previstos - std_dev

# Verificar quantas vezes a cotao de venda ultrapassa os limites (apenas no ltimo ano)
tocou_superior = df_filtrado['cotacaoVenda'] > limite_superior
tocou_inferior = df_filtrado['cotacaoVenda'] < limite_inferior
print(tocou_superior)
# Contar os toques
num_tocou_superior = tocou_superior.sum()
num_tocou_inferior = tocou_inferior.sum()

# Total de toques
total_tocou = num_tocou_superior + num_tocou_inferior

# Exibir os resultados
print(f"Quantidade de vezes que o preo tocou ou ultrapassou o desvio padro superior (ltimo ano): {num_tocou_superior}")
print(f"Quantidade de vezes que o preo tocou ou ultrapassou o desvio padro inferior (ltimo ano): {num_tocou_inferior}")
print(f"Total de toques nos desvios padro (ltimo ano): {total_tocou}")
```
